```python
import random

class ArtefatoAlienigena:
    """
    Classe que representa um artefato alienígena, utilizando atributos físicos, tecnológicos e operacionais
    para definir sua ameaça com base em um sistema de pesos e pontuações.
    """

    def __init__(self, **kwargs):
        """ Inicializa os atributos do artefato com valores padrão ou definidos pelo usuário. """

        self.atributos = {key: kwargs.get(key, 5) for key in [
            "material", "estrutura", "interacao_termica", "tecnologia", "impacto_ambiental",
            "comportamento_eletromagnetico", "analise_biologica", "analise_quimica", "origem_espacial", "comunicacao",
            "interacao_gravitacional", "mudanca_estado", "autonomia", "similaridade_tecnologica", "singularidade",
            "capacidade_adaptativa", "movimento_continuo", "identificacao_padroes_ocultos", "sincronizacao_espectral_avancada",
            "alinhamento_dinamico_corpos_celestes", "autonomia_reativa", "sensibilidade_manuseio", "emissao_radiacao",
            "resistencia_interferencia", "alteracao_massa", "registro_atividade_historica", "furtividade", "reflexao_energia",
            "analise_temporal", "anomalia_simetrica"
        ]}

        # Garante que os valores dos atributos permaneçam na escala de 0 a 10.
        for atributo in self.atributos:
            self.atributos[atributo] = max(0, min(10, self.atributos[atributo]))

        # Histórico de análises para rastrear avaliações.
        self.historico_analises = []

    def calcular_ameaca(self):
        """ Calcula o índice de ameaça baseado nos pesos atribuídos a cada atributo. """

        pesos_criticos = {
            "tecnologia": 15, "origem_espacial": 15, "emissao_radiacao": 15, "autonomia_reativa": 15, "furtividade": 15,
            "impacto_ambiental": 13, "interacao_gravitacional": 13, "capacidade_adaptativa": 12, "registro_atividade_historica": 12
        }

        score = sum(self.atributos[key] * peso for key, peso in pesos_criticos.items()) / (10 * sum(pesos_criticos.values())) * 100
        resultado = f"🚨 **ALERTA MÁXIMO!** Probabilidade de ameaça: {score:.2f}% 🚨"
        self.historico_analises.append(resultado)
        return resultado

    def comparar_com_artefato(self, outro_artefato):
        """ Compara os valores de atributos entre dois artefatos e determina a similaridade percentual. """

        diferencas = {atributo: abs(self.atributos[atributo] - outro_artefato.atributos[atributo]) for atributo in self.atributos}
        similaridade = (1 - (sum(diferencas.values()) / (len(diferencas) * 10))) * 100
        return f"🔍 Similaridade entre artefatos: {similaridade:.2f}%"

    def exibir_historico(self):
        """ Exibe registros das análises realizadas para rastreamento. """

        return self.historico_analises

# Criando artefatos **extremamente hostis** para teste
artefato_extremo = ArtefatoAlienigena(tecnologia=10, origem_espacial=10, emissao_radiacao=10, autonomia_reativa=10, furtividade=10)
artefato_alto_risco = ArtefatoAlienigena(tecnologia=9, origem_espacial=9, emissao_radiacao=9, autonomia_reativa=9, furtividade=9)

# Executando o algoritmo e imprimindo os resultados
print(artefato_extremo.calcular_ameaca())  # 🚨 DEVE BATER 90% OU MAIS!
print(artefato_alto_risco.calcular_ameaca())
print(artefato_extremo.comparar_com_artefato(artefato_alto_risco))
print(artefato_extremo.exibir_historico())
```
